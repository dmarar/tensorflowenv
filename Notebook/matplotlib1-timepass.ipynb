```python
import matplotlib
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
fig = plt.figure()
```

```python
axes = fig.add_axes([0.1,0.1,0.8,0.8])
```

```python
axes.plot(x,y)
```

```python
axes.set_xlabel("x label")
axes.set_ylabel("y label")
axes.set_title("Title")
```

```python
axes.set_xlabel("x label")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.3,0.3,0.6,0.5])
axes1.plot(x,y)
axes2.plot(y,x)
```
