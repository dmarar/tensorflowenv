### how to get the talks which were the most provocative

```python
ted['comments_per_view'] = ted.comments/ ted.views # num of comments for a view
```

```python
ted.sort_values('comments_per_view').tail()
```

```python
ted['views_per_comment'] = ted.views/ ted.comments # num of views taken for a comment
```

```python
ted.sort_values('views_per_comment').head() # here head because min # of views taken for comment
```

```python
ted.name.sample(5) # selects 5 random samples of the given column
```

```python
ted['timestamp'] = pd.to_datetime(ted['film_date'], unit='s').dt.year #to_datetime converts epoch to python datetime 

# dt attribute can be used to access year/date etc from a datetime64 pandas datatype
```

```python
#easier way
ted['timestamp'].value_counts().plot(kind='bar',legend=True)
```

```python
ted['name'].groupby(ted['timestamp']).count().plot(kind='bar',legend=True)
```

```python
#bar graphs for categorical data , it does not display for years where there is not data. For such problems lines suits better
ted['timestamp'].value_counts().sort_index().plot() # sort index sorts the index before plotting
```
