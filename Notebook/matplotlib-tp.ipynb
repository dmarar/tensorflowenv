```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
gas = pd.read_csv('gas_prices.csv')
```

```python
gas.head()
```

```python
plt.title('gas prices across countries')
plt.xlabel('Year')
plt.ylabel('Gallons')

plt.plot(gas.Year, gas.USA,label='USA')
plt.plot(gas.Year, gas.Japan,label='Japan')
plt.plot(gas.Year, gas.UK,label='UK')
plt.legend()
```

```python
# gas.Year[::3]
plt.figure(figsize=(15,6))
plt.xticks(gas.Year[::3])
for country in gas:
    if country!= 'Year':
#         print(country)
        plt.plot(gas.Year, gas[country], label=country)
    
plt.legend()
```
