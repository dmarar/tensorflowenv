```python
from yellowbrick.target import FeatureCorrelation
feature_name = list(features.columns)
```

```python
 plt.figure(figsize=(10,8))
 boston_corr = boston_data.corr()
sns.heatmap(boston_corr, linewidth=1,annot=True, annot_kws={'size':10})
# boston_data.corr()
```

```python
visualizer = FeatureCorrelation(labels=feature_name)
visualizer.fit(features, target)
visualizer.show()
```

```python
from sklearn.feature_selection import SelectKBest, f_regression
univariate = SelectKBest(k=5, score_func=f_regression).fit(features, target)
mask = univariate.get_support()
features.columns[mask]
```

```python
univariate.scores_
```

```python
feature_df = pd.DataFrame({'feature': features.columns, 'scores': univariate.scores_}).sort_values(by='scores', ascending=False)
```

```python
uni_df = pd.DataFrame({'Univariate method': features.columns[mask]})
uni_df
```

```python
#wrapper methods using forward and backward 
from sklearn.feature_selection import RFE # RFE Recursive feature elimination
from sklearn.linear_model import LinearRegression
rfe = RFE(estimator=LinearRegression(), n_features_to_select=5)
rfe.fit(features, target)
```

```python
#get only those features from the rfe estimator
rfe_features = features.columns[rfe.support_]
rfe_features
# features.columns[rfe_features]
rfe_df = pd.DataFrame({'RFE Method': rfe_features})
rfe_df
```
